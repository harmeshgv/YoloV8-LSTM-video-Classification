```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
non_violence_data = pd.read_csv("../non_violence_data.csv")
violence_data = pd.read_csv("../violence_data.csv")

```

```python
non_violence_data
```

```python
violence_data["label"] = 1
non_violence_data["label"] = 0
```

```python
data = pd.concat([violence_data, non_violence_data], ignore_index=True)
```

```python
X = data.drop(columns=[ 'frame_index','timestamp','person1_id', 'person2_id', 'person1_kp0_x', 'person1_kp0_y',
 'person1_kp1_x',
 'person1_kp1_y',
 'person1_kp2_x',
 'person1_kp2_y',
 'person1_kp3_x',
 'person1_kp3_y',
 'person1_kp4_x',
 'person1_kp4_y','person2_kp0_x',
 'person2_kp0_y',
 'person2_kp1_x',
 'person2_kp1_y',
 'person2_kp2_x',
 'person2_kp2_y',
 'person2_kp3_x',
 'person2_kp3_y',
 'person2_kp4_x',
 'person2_kp4_y','relative_kp0_x',
 'relative_kp0_y',
 'relative_kp1_x',
 'relative_kp1_y',
 'relative_kp2_x',
 'relative_kp2_y',
 'relative_kp3_x',
 'relative_kp3_y',
 'relative_kp4_x',
 'relative_kp4_y','label'])
```

```python
Y = data['label']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_val, y_train, y_val = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.preprocessing import StandardScaler
```

```python
SS = StandardScaler()
X = SS.fit_transform(X)
```

```python
from sklearn.linear_model import LogisticRegression
log_params = {
    "C": [0.01, 1, 100],                 # Small, medium, large regularization
    "penalty": ["l1", "l2"],             # L1 (sparse), L2 (default, stable)
    "solver": ["liblinear", "saga"],     # Supports L1 + L2
    "max_iter": [1000],                  # High enough to ensure convergence
    "class_weight": [None, "balanced"]   # Handle imbalance
}


log_model = LogisticRegression()  

grid_search = GridSearchCV(log_model, log_params, cv = 10, scoring='accuracy', verbose=2, n_jobs=-1)
grid_search.fit(X, Y)

print(f"Best Params: {grid_search.best_params_}")
print(f"Best Score: {grid_search.best_score_}")
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

logestic = LogisticRegression(C= 0.01, class_weight= 'balanced', max_iter = 1000, penalty = 'l1', solver = 'liblinear')
logestic.fit(x_train, y_train)
y_pred = logestic.predict(x_val)

print("Accuracy:", accuracy_score(y_val, y_pred))
print("Confusion Matrix:\n", confusion_matrix(y_val, y_pred))
print("Classification Report:\n", classification_report(y_val, y_pred))
```

```python
from sklearn.tree import DecisionTreeClassifier

model = DecisionTreeClassifier(
    criterion='gini',        
    max_depth=None,          
    class_weight='balanced',
    random_state=42
)

```

```python
model.fit(x_train, y_train)
y_pred = model.predict(x_val)
```

```python
print("Accuracy:", accuracy_score(y_val, y_pred))
print("Confusion Matrix:\n", confusion_matrix(y_val, y_pred))
print("Classification Report:\n", classification_report(y_val, y_pred))
```

```python
import os 
import pandas as pd
from tqdm import tqdm
from IPython.display import clear_output    
```

```python
sample_files= ["t_n001_converted.avi"]
sample_file_path = ".."
```

```python
import sys
import os

# go one step up to project root
sys.path.append(os.path.abspath(".."))
```

```python
from backend.pipeline import ViolenceFeatureExtractor
from backend.utils.data_preprocessing import preprocess_data

# Absolute paths to your models
detection_model_path = "backend/models/yolo8n.pt"
pose_model_path = "backend/models/yolo8n.pt"

# Check if the model files exist
if not os.path.exists(detection_model_path):
    raise FileNotFoundError(f"Detection model not found at {detection_model_path}")

if not os.path.exists(pose_model_path):
    raise FileNotFoundError(f"Pose model not found at {pose_model_path}")

# Initialize the feature extractor
extractor = ViolenceFeatureExtractor()

# Path to the final CSV where data will be appended
final_csv_path = 'sample.csv'

# Path to save the output CSV
output_csv_path = 'extracted_feature_data/sample.csv'

full_log = []

# tqdm loop with description
for video_name in tqdm(sample_files, desc="Processing Videos"):
    try:
        extractor.reset()
        video_path = os.path.join(sample_file_path, video_name)

        print(f"\nüìπ Processing video: {video_name}")
        frame_width, frame_height, interaction_count = extractor.process_video(
            video_path,
            output_csv_path,
            output_folder=None,
            show_video=False,
            save_video=False
        )
        clear_output(wait=True)

        df_csv = pd.read_csv(output_csv_path)
        df_preprocessed, expected_columns = preprocess_data(df_csv, frame_width, frame_height)

        # Skip appending if there are no interactions (empty dataframe)
        if df_preprocessed.empty:
            log_entry = f"‚ùå {video_name}: No interactions found"
            print(log_entry)
            full_log.append(log_entry)
            continue  # Skip to the next video

        # Check if the final CSV exists and has data
        if os.path.exists(final_csv_path) and os.path.getsize(final_csv_path) > 0:
            # Append to the existing CSV
            df_preprocessed.to_csv(final_csv_path, mode='a', header=False, index=False)
        else:
            # Create a new CSV with the appropriate header
            df_preprocessed.to_csv(final_csv_path, mode='w', header=True, index=False)

        log_entry = f"‚úÖ {video_name}: {len(df_preprocessed)} interactions saved"
        print(log_entry)

    except Exception as e:
        log_entry = f"‚ùå Error processing {video_name}: {e}"
        print(log_entry)

    full_log.append(log_entry)

```
