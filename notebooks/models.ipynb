```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
train_violence_data = pd.read_csv("extracted_feature_data/train_violence_data.csv")
train_non_violence_data = pd.read_csv("extracted_feature_data/train_non_violence_data.csv")

val_violence_data = pd.read_csv("extracted_feature_data/val_violence_data.csv")
val_non_violence_data = pd.read_csv("extracted_feature_data/val_non_violence_data.csv")
```

```python
train_violence_data["label"] = 1
val_violence_data["label"] = 1
train_non_violence_data["label"] = 0
val_non_violence_data["label"] = 0
```

```python
final_train_data = pd.concat([train_violence_data, train_non_violence_data], ignore_index=True)
final_val_data = pd.concat([val_violence_data, val_non_violence_data], ignore_index=True)
```

```python
final_train_data
```

```python
combine_data = pd.concat([final_train_data, final_val_data], ignore_index=True)
```

```python
X = combine_data.drop(columns=[ 'video_name','frame_index','timestamp','person1_id', 'person2_id', 'person1_kp0_x', 'person1_kp0_y',
 'person1_kp1_x',
 'person1_kp1_y',
 'person1_kp2_x',
 'person1_kp2_y',
 'person1_kp3_x',
 'person1_kp3_y',
 'person1_kp4_x',
 'person1_kp4_y','person2_kp0_x',
 'person2_kp0_y',
 'person2_kp1_x',
 'person2_kp1_y',
 'person2_kp2_x',
 'person2_kp2_y',
 'person2_kp3_x',
 'person2_kp3_y',
 'person2_kp4_x',
 'person2_kp4_y','relative_kp0_x',
 'relative_kp0_y',
 'relative_kp1_x',
 'relative_kp1_y',
 'relative_kp2_x',
 'relative_kp2_y',
 'relative_kp3_x',
 'relative_kp3_y',
 'relative_kp4_x',
 'relative_kp4_y','label'])
```

```python
Y = combine_data['label']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_val, y_train, y_val = train_test_split(X, Y, test_size=0.2, random_state=42, stratify=Y)
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(class_weight='balanced', max_iter=1000)  
model.fit(x_train, y_train)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

y_pred = model.predict(x_val)

print("Accuracy:", accuracy_score(y_val, y_pred))
print("Confusion Matrix:\n", confusion_matrix(y_val, y_pred))
print("Classification Report:\n", classification_report(y_val, y_pred))
```

```python
from sklearn.tree import DecisionTreeClassifier

model = DecisionTreeClassifier(
    criterion='gini',        
    max_depth=None,          
    class_weight='balanced',
    random_state=42
)

```

```python
model.fit(x_train, y_train)
y_pred = model.predict(x_val)
```

```python
print("Accuracy:", accuracy_score(y_val, y_pred))
print("Confusion Matrix:\n", confusion_matrix(y_val, y_pred))
print("Classification Report:\n", classification_report(y_val, y_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
import os 
import pandas as pd
from tqdm import tqdm
from IPython.display import clear_output    
```

```python
sample_files= ["V_4.mp4"]
sample_file_path = "video_data/violence"
```

```python

from backend.utils.feature_extraction import ViolenceFeatureExtractor
from backend.utils.data_preprocessing import preprocess_data

# Absolute paths to your models
detection_model_path = "backend/models/yolo8n.pt"
pose_model_path = "backend/models/yolo8n.pt"

# Check if the model files exist
if not os.path.exists(detection_model_path):
    raise FileNotFoundError(f"Detection model not found at {detection_model_path}")

if not os.path.exists(pose_model_path):
    raise FileNotFoundError(f"Pose model not found at {pose_model_path}")

# Initialize the feature extractor
extractor = ViolenceFeatureExtractor()

# Path to the final CSV where data will be appended
final_csv_path = 'sample.csv'

# Path to save the output CSV
output_csv_path = 'extracted_feature_data/sample.csv'

full_log = []

# tqdm loop with description
for video_name in tqdm(sample_files, desc="Processing Videos"):
    try:
        extractor.reset()
        video_path = os.path.join(sample_file_path, video_name)

        print(f"\n📹 Processing video: {video_name}")
        frame_width, frame_height, interaction_count = extractor.process_video(
            video_path,
            output_csv_path,
            output_folder=None,
            show_video=False,
            save_video=False
        )
        clear_output(wait=True)

        df_csv = pd.read_csv(output_csv_path)
        df_preprocessed, expected_columns = preprocess_data(df_csv, frame_width, frame_height)

        # Skip appending if there are no interactions (empty dataframe)
        if df_preprocessed.empty:
            log_entry = f"❌ {video_name}: No interactions found"
            print(log_entry)
            full_log.append(log_entry)
            continue  # Skip to the next video

        # Check if the final CSV exists and has data
        if os.path.exists(final_csv_path) and os.path.getsize(final_csv_path) > 0:
            # Append to the existing CSV
            df_preprocessed.to_csv(final_csv_path, mode='a', header=False, index=False)
        else:
            # Create a new CSV with the appropriate header
            df_preprocessed.to_csv(final_csv_path, mode='w', header=True, index=False)

        log_entry = f"✅ {video_name}: {len(df_preprocessed)} interactions saved"
        print(log_entry)

    except Exception as e:
        log_entry = f"❌ Error processing {video_name}: {e}"
        print(log_entry)

    full_log.append(log_entry)

```
